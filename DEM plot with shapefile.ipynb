```python
import rasterio
import rasterio.plot
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
```

```python
raster_path = ("DEM mask.tif")
src = rasterio.open(raster_path)
print(src)
```

```python
##Basic Raster Plot
rasterio.plot.show(src)
```

```python
##customize the plot
fig, ax = plt.subplots(figsize=(8, 8))
rasterio.plot.show(src, cmap="terrain", ax=ax, title="Digital Elevation Model (DEM)")
plt.show()
```

```python
##ploting vcetor on raster
dem_bounds = ("E:/Pradip/east coast shape file/east_coast.shp")
gdf = gpd.read_file(dem_bounds)
gdf = gdf.to_crs(src.crs)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
rasterio.plot.show(src, cmap="terrain", ax=ax, title="Digital Elevation Model (DEM)")
gdf.plot(ax=ax, edgecolor="red", facecolor="none", lw=0.5)
```
